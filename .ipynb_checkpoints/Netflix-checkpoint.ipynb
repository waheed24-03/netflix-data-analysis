```python
import pandas as pd
```

```python
df=pd.read_csv("netflix_titles.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df['date_added'] = pd.to_datetime(df['date_added'].astype(str).str.strip(), errors='coerce')
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

```python
df['type'].value_counts()
```

## Number of Movies vs TV Shows

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='type', palette='Set2')
plt.title("Number of Movies vs TV Shows")
plt.show()
```

## Content Added Each Year


```python
df['year_added'].value_counts().sort_index().plot(kind='bar', figsize=(10,4), color='orange')
plt.title("Content Added to Netflix by Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.tight_layout()
plt.show()
```

## Top 10 Countries by Netflix Content

```python
df['country'].value_counts().head(10).plot(kind='bar',color='teal')
plt.title("Top 10 countries with Most Content on Netflix")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```

##  Genre Analysis

```python
from collections import Counter
genre_series=df['listed_in'].dropna().apply(lambda x: x.split(', '))
all_genres=Counter([genre for sublist in genre_series for genre in sublist])
top_genres=pd.DataFrame(all_genres.most_common(10),columns=['Genre','Count'])
sns.barplot(data=top_genres,x='Count', y='Genre', palette='rocket')
plt.title("Top 10 Most Common Genres on Netfllix")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.show()
```

## Top Actors on Netflix


```python
actor_series=df['cast'].dropna().apply(lambda x: x.split(', '))
all_actors=Counter([actor.strip() for sublist in actor_series for actor in sublist])
top_actors=pd.DataFrame(all_actors.most_common(10), columns=['Actor', 'Appearances'])
sns.barplot(data=top_actors, x='Appearances', y='Actor', palette='mako')
plt.title("Top10 Most Featured Actors on Neflix")
plt.xlabel("Number of Titles")
plt.ylabel("Actor")
plt.show()
```

## Highest Rated Shows & Movies

```python
df['rating'].value_counts().head(10).plot(kind='bar', color='purple')
plt.title("Top Content Ratings on Neflix")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
rating_order = [
    "G", "TV-Y", "TV-G", 
    "PG", "TV-PG", 
    "PG-13", "TV-14", 
    "R", "TV-MA", "NC-17"
]

```

##  Find Best Movie/Show for Each Country

```python
top_countries=['United States', 'India', 'United Kingdom', 'Japan', 'France']
df['rating_rank']=df['rating'].apply(lambda x: rating_order.index(x) if x in rating_order else -1)
best_per_country=[]
for country in top_countries:
    country_df=df[df['country'].str.contains(country, na=False)]
    best= (    
        country_df[country_df['rating_rank'] != -1]
        .sort_values(['rating_rank', 'release_year'], ascending=[False, False])
        .head(1)
    )
    best_per_country.append(best)
best_df=pd.concat(best_per_country)
```

```python
from IPython.display import display, HTML

def make_card(row):
    return f"""
    <div style="background:#f2f2f2; border-radius:10px; padding:15px; margin:10px 0">
        <h3>{row['title']} <small style='font-size:14px'>({int(row['release_year'])})</small></h3>
        <p><strong>Country:</strong> {row['country']}<br>
        <strong>Type:</strong> {row['type']}<br>
        <strong>Rating:</strong> {row['rating']}<br>
        <strong>Genre:</strong> {row['listed_in']}</p>
    </div>
    """

HTML("".join([make_card(row) for _, row in best_df.iterrows()]))
```
